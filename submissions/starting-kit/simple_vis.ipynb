```python
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path = '../../data/train/full_train_data.csv'
df = pd.read_csv(path)
df.head()
```

```python
plt.figure(figsize=(15, 8))
ax = sns.countplot(x="Type 1", hue="Legendary", data=df)

plt.xticks(rotation= 90)
plt.xlabel('Type 1')
plt.ylabel('Total ')
plt.title("Occurence avec LÃ©gendaire pour le Type 1")
```

```python
plt.figure(figsize=(15, 8))
ax = sns.countplot(x="Type 2", hue="Legendary", data=df)

plt.xticks(rotation= 90)
plt.xlabel('Type 2')
plt.ylabel('Total ')
plt.title("Occurence avec LÃ©gendaire pour le Type 2")
```

```python
# Carte de corrÃ©lation 
plt.figure(figsize=(10, 8))
plt.title('CorrÃ©alation des variables')


matrix = df.drop(columns=['#']).corr().round(2)
sns.heatmap(matrix, annot=True)
plt.show()
```
