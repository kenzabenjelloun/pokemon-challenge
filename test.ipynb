```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
data = pd.read_csv("data/initial_data/pokemon.csv")
data
```

```python
test_df = pd.DataFrame({})
test_df['Type 1'] = data['Type 1']
test_df['Type 2'] = data['Type 2']
```

```python
test_df['Type 1'].values
```

```python
data = test_df['Type 1'].values.reshape(-1, 1)

one_hot = OneHotEncoder(handle_unknown='ignore')
one_hot.fit(data)
```

```python
type_two_data = test_df['Type 2'].values.reshape(-1, 1)
```

```python
res = pd.DataFrame({})

type_one_data = test_df['Type 1'].values.reshape(-1, 1)
type_two_data = test_df['Type 2'].values.reshape(-1, 1)
res['Type 1'] = one_hot.transform(type_one_data)
res['Type 2'] = one_hot.transform(type_two_data)

res
```
