### Notebook pour la fusion de "pokemon.csv" et "data.csv"

```python
import pandas as pd
import numpy as np
```

```python
combat_df = pd.read_csv('combats.csv')
poke_df = pd.read_csv('pokemon.csv')
```

#### Fusion des données

```python
nb_wins  = combat_df['Winner'].value_counts().sort_index()
nb_combats = combat_df['First_pokemon'].value_counts().sort_index() + combat_df['Second_pokemon'].value_counts().sort_index()

win_ratio = nb_wins / nb_combats
win_ratio
```

```python
df = pd.concat([poke_df.set_index('#'), win_ratio], axis=1).reset_index()
df = df.rename(columns={'index': '#', 0: 'Win_ratio'})
df
```

```python
df.to_csv('full_train_data.csv', index=False)
```
